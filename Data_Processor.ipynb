```python
import pandas as pd
from datetime import timedelta, date
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
start_date = date(2016, 8, 1)
end_date = date(2017, 8, 2)
```

```python
def dataToCsv(single_date):
    filename = single_date.strftime("%Y%m%d") + ".json"
    data = pd.read_json("Group/" + filename)
    deviceCategory_Column = data.device.apply(pd.Series).deviceCategory
    totalTransactionRevenue_Column = data.totals.apply(pd.Series).totalTransactionRevenue.fillna(0) / 10**6
    isTrsacationRevenue_Column = totalTransactionRevenue_Column.clip(upper=1)
    isTrsacationRevenue_Column.name = "isTrsacationRevenue"
    channelGrouping_Column = data.channelGrouping
    # remember to drop the '(not set)' rows
    continent_Column = data.geoNetwork.apply(pd.Series).continent
    bounces_Column = data.totals.apply(pd.Series).bounces.fillna(0)
    date_Column = data.date
    month_Column = date_Column // 100
    month_Column.name = 'month'
    result = pd.concat([month_Column, date_Column, deviceCategory_Column,totalTransactionRevenue_Column, isTrsacationRevenue_Column, channelGrouping_Column, continent_Column, bounces_Column], axis=1, sort=False)
    result = result.reset_index(drop=True)
    result = result[result["continent"]!="(not set)"]
    result.to_csv("projectData/" + single_date.strftime("%Y%m%d") + ".csv")
    print("Date: " + single_date.strftime("%Y%m%d") + " finished")
```

```python
dates = daterange(start_date, end_date)
```

```python
def resample(single_date):
    filename = single_date.strftime("%Y%m%d") + ".csv"
    data = pd.read_csv("projectData/" + filename)
    data_to_add = data[data['isTrsacationRevenue'] == 1]
    data_to_concat = pd.concat([data_to_add] * 49)
    final_data = data.append(data_to_concat)
    final_data.to_csv("resampledData/" + single_date.strftime("%Y%m%d") + ".csv")
    print("Date: " + single_date.strftime("%Y%m%d") + " finished")
```
